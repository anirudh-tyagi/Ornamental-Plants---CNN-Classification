```python
base_path="Ornamental_Plants"
```

```python
damask_rose = glob.glob(os.path.join(base_path, "Damask Rose", "*.jp*g")) + glob.glob(os.path.join(base_path, "Damask Rose", "*.png"))
rain_lily = glob.glob(os.path.join(base_path, "Rain Lily", "*.jp*g")) + glob.glob(os.path.join(base_path, "Rain Lily", "*.png"))
echeveria_flower = glob.glob(os.path.join(base_path, "Echeveria Flower", "*.jp*g")) + glob.glob(os.path.join(base_path, "Echeveria Flower", "*.png"))
mirabilis_jalapa = glob.glob(os.path.join(base_path, "Mirabilis Jalapa", "*.jp*g")) + glob.glob(os.path.join(base_path, "Mirabilis Jalapa", "*.png"))
zinnia_elegans = glob.glob(os.path.join(base_path, "Zinnia Elegans", "*.jp*g")) + glob.glob(os.path.join(base_path, "Zinnia Elegans", "*.png"))
```

```python
print(f"Damask Rose: {len(damask_rose)} files")
print(f"Rain Lily: {len(rain_lily)} files")
print(f"Echeveria Flower: {len(echeveria_flower)} files")
print(f"Mirabilis Jalapa: {len(mirabilis_jalapa)} files")
print(f"Zinnia Elegans: {len(zinnia_elegans)} files")
```

```python
total_drose = len(damask_rose)
total_rlily = len(rain_lily)
total_eflower = len(echeveria_flower)
total_mjalapa = len(mirabilis_jalapa)
total_zelegans = len(zinnia_elegans)

imgFiles = (
    damask_rose[:round(0.4*total_drose)] +
    rain_lily[:round(0.4*total_rlily)] +
    echeveria_flower[:round(0.4*total_eflower)] +
    mirabilis_jalapa[:round(0.4*total_mjalapa)] +
    zinnia_elegans[:round(0.4*total_zelegans)]
)

# Print first 5 file paths
print("First 5 image files:")
for items in imgFiles[:5]:
    print(items)

# Print statistics
print("\nDataset Statistics:")
print(f"Total number of images: {len(imgFiles)}")
print(f"Damask Rose images: {round(0.4*total_drose)}")
print(f"Rain Lily images: {round(0.4*total_rlily)}")
print(f"Echeveria Flower images: {round(0.4*total_eflower)}")
print(f"Mirabilis Jalapa images: {round(0.4*total_mjalapa)}")
print(f"Zinnia Elegans images: {round(0.4*total_zelegans)}")
```

```python
X = []
y = []

for fName in imgFiles:

  # Prepare the dataset and populate X and y
  X_i = Image.open(fName)
  X_i = X_i.resize((299,299))
  if X_i.mode != 'RGB':
    X_i = X_i.convert('RGB')
  X_i = np.array(X_i) / 255.0

  X.append(X_i)

  label = fName.split("/")
  y_i = label[-2]

  y.append(y_i)
```

```python

from sklearn.preprocessing import LabelEncoder

lEncoder = LabelEncoder()
y = lEncoder.fit_transform(y)

print(set(y))
print(lEncoder.classes_)
from sklearn.preprocessing import LabelEncoder

lEncoder = LabelEncoder()
y = lEncoder.fit_transform(y)

print(set(y))
print(lEncoder.classes_)
```

```python
X = np.array(X)
y = np.array(y)

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,
                                                    stratify=y, random_state=42)

print("X_train_shape: {}".format(X_train.shape))
print("X_test_shape: {}".format(X_test.shape))
```

```python
mu = X_train.mean()
std = X_train.std()

X_train_std = (X_train-mu)/std # Apply feature scaling to X_train
X_test_std = (X_test-mu)/std # Apply feature scaling to X_test
```
